# 03. Creating new tables to explore relationships

#### In order to explore relationships between different values, I'll create 3 different tables:

#### - Units sold: this table will aggregate sales per product id and include different product characteristics such as the number of pictures uploaded, the length of the title and the length of the description.

#### - Payment preferences: this table will aggregate the orders with the amount, number of installments and count of payment methods.

#### - Order reviews: this table will aggregate the orders by rating and delivery time.

## Units sold

```python
# Create new dataset with an aggregate of sales and different product characteristics

units_sold = df.groupby('product_id').agg({'order_product_item_count': ['nunique'],'product_photos_qty': ['mean'],'product_name_length': ['mean'],'product_description_length': ['mean']})
```

## Order reviews

```python
# Create new column to measure how long it takes Olist to fullfill customers' orders

df['order_lead_time'] = df['order_actual_delivery_date'] - df['timestamp_order']

# Convert the difference in days

df['order_lead_time'] = df['order_lead_time'] / np.timedelta64(1,'D')

# Check output
df
```

```python
# Create new column to measure how long it takes Olist to fullfill customers' orders

df['order_days_delayed'] = df['order_actual_delivery_date'] - df['order_estimated_delivery_date']

# Convert the difference in days

df['order_days_delayed'] = df['order_days_delayed'] / np.timedelta64(1,'D')

# Check output
df
```

```python
# Create a subset for delivered orders

df_delivered =  df[df['order_status']=='delivered']
```

```python
# Create new dataset

order_reviews = df_delivered.groupby('order_id').agg({'review_score': ['mean'],'order_lead_time': ['mean'],'order_days_delayed': ['mean']})
```

```python
# check dataset

order_reviews.head()
```

```python
# Export new dataset

order_reviews.to_csv(os.path.join(path, '02 Data','Prepared Data', 'order_reviews.csv'))
```
