## Dropping columns

```python
# Drop columns from merged dataset

df_wrangled = df_merged.drop(columns = ['Unnamed: 0','product_category_name'])
```

## Renaming columns

```python
# Change name of column "order_dow" to orders_day_of_week

df_wrangled.rename(columns = {'freight_value' : 'item_shipping_cost'}, inplace = True)
df_wrangled.rename(columns = {'order_item_id' : 'order_item_count'}, inplace = True)
df_wrangled.rename(columns = {'order_purchase_timestamp' : 'timestamp_order'}, inplace = True)
df_wrangled.rename(columns = {'order_approved_at' : 'timestamp_payment_approval'}, inplace = True)
df_wrangled.rename(columns = {'order_delivered_carrier_date' : 'timestamp_order_dispatched'}, inplace = True)
df_wrangled.rename(columns = {'order_delivered_customer_date' : 'order_actual_delivery_date'}, inplace = True)
df_wrangled.rename(columns = {'order_estimated_delivery_date' : 'order_estimated_delivery_date'}, inplace = True)
df_wrangled.rename(columns = {'payment_sequential' : 'payment_method_count'}, inplace = True)
df_wrangled.rename(columns = {'review_comment_title' : 'review_title'}, inplace = True)
df_wrangled.rename(columns = {'review_comment_message' : 'review_text'}, inplace = True)
df_wrangled.rename(columns = {'review_creation_date' : 'survey_sent_date'}, inplace = True)
df_wrangled.rename(columns = {'review_answer_timestamp' : 'survey_submision_date'}, inplace = True)
df_wrangled.rename(columns = {'payment_value' : 'order_amount'}, inplace = True)
df_wrangled.rename(columns = {'price' : 'item_price'}, inplace = True)
df_wrangled.rename(columns = {'product_name_lenght' : 'product_name_length'}, inplace = True)
df_wrangled.rename(columns = {'product_description_lenght' : 'product_description_length'}, inplace = True)
```

```python
# Check output

df_wrangled.info()
```

## Changing data types

```python
# Change types for data set

df_wrangled['seller_zip_code_prefix']=df_wrangled['seller_zip_code_prefix'].astype('object')
df_wrangled['customer_zip_code_prefix']=df_wrangled['customer_zip_code_prefix'].astype('object')
df_wrangled['order_actual_delivery_date'] = df_wrangled['order_actual_delivery_date'].astype('datetime64[ns]')
df_wrangled['order_estimated_delivery_date'] = df_wrangled['order_estimated_delivery_date'].astype('datetime64[ns]')
df_wrangled['timestamp_order'] = df_wrangled['timestamp_order'].astype('datetime64[ns]')
df_wrangled['timestamp_payment_approval'] = df_wrangled['timestamp_payment_approval'].astype('datetime64[ns]')
df_wrangled['timestamp_order_dispatched'] = df_wrangled['timestamp_order_dispatched'].astype('datetime64[ns]')
df_wrangled['product_name_length']=df_wrangled['product_name_length'].astype('int64')
df_wrangled['product_description_length']=df_wrangled['product_description_length'].astype('int64')
df_wrangled['product_photos_qty']=df_wrangled['product_photos_qty'].astype('int64')
```

## Checking for Mixed-Type Data

```python
# Check for mixed types

for col in df_wrangled.columns.tolist():
  weird = (df_wrangled[[col]].applymap(type) != df_wrangled[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_wrangled[weird]) > 0:
    print (col)
```

### The mixed types are because there are missing values in these columns. I'll convert to strings only the review title and text.

```python
# Convert to string

df_wrangled['review_title'] = df_wrangled['review_title'].astype('str')
df_wrangled['review_text'] = df_wrangled['review_text'].astype('str')
```
